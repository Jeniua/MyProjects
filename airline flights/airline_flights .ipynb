# Спрос на рейсы авиакомпании

## Аналитика

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
query_1 = pd.read_csv('/datasets/query_1.csv')
print(query_1.info())
```

```python
query_3 = pd.read_csv('/datasets/query_3.csv')
print(query_3.info())
```

```python
top_city = query_3.sort_values('average_flights', ascending = False)
print(top_city.head(10))
```

```python
plt.figure(figsize=(20,6))
plt.bar(top_city.city, top_city.average_flights)
plt.title('Среднее кол-во рейсов, прибывающих за день в различные города РФ')
plt.ylabel('Прибывающие рейсы')
plt.xlabel('Город')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10 = top_city.head(10)
plt.bar(top_10.city, top_10.average_flights)
plt.title('Среднее кол-во рейсов, прибывающих за день в популярные города РФ')
plt.ylabel('Прибывающие рейсы')
plt.xlabel('Город')
plt.xticks(rotation=90)
plt.show()
```

```python
query_1.sort_values(by='flights_amount', ascending=False)
```

```python
query_1 = query_1.sort_values(by='flights_amount', ascending=False)
plt.figure(figsize=(20,6))
plt.bar(query_1.model, query_1.flights_amount)
plt.title('Количество полетов различных моделей самолетов')
plt.ylabel('Количество рейсов')
plt.xlabel('Модель самолета')
plt.show()
```

### Вывод
По полученным данным, можно сказать, что наиболее популярным городом по количеству рейсов является Москва (как и ожидалось, столица пользуется спросом).  
Самолеты моделей Cessna 208 Caravan, Bombardier CRJ-200, Sukhoi SuperJet-100 выполняют больше всего рейсов. Их показатели примерно равны.
